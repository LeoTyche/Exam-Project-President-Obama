```python
dfnew = dfnew.rename(columns={'image_url': 'url'})
dfnew
```

```python
dflong = pd.concat([dfnew, generated[['url']]], axis=0)
dflong = dflong.reset_index(drop=True)
dflong = dflong.fillna(1)
dflong
```

```python
import io
import requests
from PIL import Image
import numpy as np
from skimage import color
from scipy.stats import skew, kurtosis


def display_artwork(url):
    response = requests.get(url)
    image = Image.open(io.BytesIO(response.content))
    lab = color.rgb2lab(image)
    # Compute the mean and standard deviation of the L, a, and b channels
    l_med, l_std = np.median(lab[:,:,0]), np.std(lab[:,:,0])
    a_med, a_std = np.median(lab[:,:,1]), np.std(lab[:,:,1])
    b_med, b_std = np.median(lab[:,:,2]), np.std(lab[:,:,2])
    # Compute the skewness and kurtosis of the L, a, and b channels
    l_skew, l_kurt = skew(lab[:,:,0].flatten()),kurtosis(lab[:,:,0].flatten())
    a_skew, a_kurt = skew(lab[:,:,1].flatten()),kurtosis(lab[:,:,1].flatten())
    b_skew, b_kurt = skew(lab[:,:,2].flatten()),kurtosis(lab[:,:,2].flatten())
    # Combine the color features into a single feature vector
    return l_med, l_std, l_skew, l_kurt, a_med, a_std, a_skew, a_kurt, b_med, b_std, b_skew, b_kurt
```

```python
dfnum = pd.DataFrame(columns=['l_med', 'l_std', 'l_skew', 'l_kurt', 'a_med', 'a_std', 'a_skew', 'a_kurt', 'b_med', 'b_std', 'b_skew', 'b_kurt'])

# Loop over each image in the directory
for index, value in dflong.iloc[0:160,0].iteritems():
    # Load and compute the image
    l_med, l_std, l_skew, l_kurt, a_med, a_std, a_skew, a_kurt, b_med, b_std, b_skew, b_kurt = display_artwork(value)

    # Add the results to the dataframe
    dfnum = dfnum.append({'l_med':l_med, 'l_std':l_std, 'l_skew':l_skew, 'l_kurt':l_kurt, 'a_med':a_med, 'a_std':a_std, 'a_skew':a_skew, 'a_kurt':a_kurt, 'b_med':b_med, 'b_std':b_std, 'b_skew':b_skew, 'b_kurt':b_kurt}, ignore_index=True)
dfnum    
```
